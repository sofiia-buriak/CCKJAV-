```python
import pandas as pd
import matplotlib.pyplot as plt

# 1.  
df = pd.read_csv(r"C:\Users\HOME\Desktop\NASA\CCKJAV-\preds.csv")

# 2.   ( _class)
class_counts = df['pred_class'].value_counts()
plt.figure(figsize=(6,4))
class_counts.plot(kind='bar', color=['#66c2a5','#fc8d62','#8da0cb'])
plt.title("  ")
plt.ylabel(" '")
plt.xlabel("")
plt.xticks(rotation=0)
plt.show()

# 3.    5 '  
sample = df.head(5)[['pred_class','prob_CONFIRMED','prob_CANDIDATE','prob_FALSE POSITIVE']]

#  stacked bar chart
sample_plot = sample.set_index(sample.index)[['prob_CONFIRMED','prob_CANDIDATE','prob_FALSE POSITIVE']]
sample_plot.plot(kind='bar', stacked=True, figsize=(8,4), colormap="Set2")
plt.title("   5 '")
plt.ylabel("")
plt.xlabel("'")
plt.legend(title="")
plt.show()

# 4. ()      
plt.figure(figsize=(6,4))
class_counts.plot(kind='bar', color=['#66c2a5','#fc8d62','#8da0cb'])
plt.title("  ")
plt.savefig("class_distribution.png", dpi=200, bbox_inches="tight")

sample_plot.plot(kind='bar', stacked=True, figsize=(8,4), colormap="Set2")
plt.title("   5 '")
plt.savefig("sample_probs.png", dpi=200, bbox_inches="tight")
```
